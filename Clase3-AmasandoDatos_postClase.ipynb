```python
def tarifa_vector(duracion):
        if duracion <= 30:
            costo = 0
        elif duracion <= 60:
            costo = 2
        elif duracion <= 90:
            costo = 6
        else:
            media_horas = (duracion-90)//30 +1
            costo = 6 + 8*media_horas
        return costo
```

```python
def tarifa(x):
        duracion = x['duration']
        tipo = x['subscription_type']
        if duracion <= 30:
            costo = 0
        elif duracion <= 60:
            costo = 2
        elif duracion <= 90:
            costo = 6
        else:
            media_horas = (duracion-90)//30 +1
            costo = 6 + 8*media_horas
        
        if costo > 100:
            costo = 100
        if tipo == 'Casual':
            return costo
        else:
            return 0.75*costo
```

Calculemos la distancia entre dos puntos.

```python
from math import sin, cos, sqrt, atan2, radians

def computeDistance(lat_x, lon_x, lat_y, lon_y):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat_x)
    lon1 = radians(lon_x)
    lat2 = radians(lat_y)
    lon2 = radians(lon_y)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
computeDistance(52.2296756,21.0122287,52.406374, 16.9251681)
```

```python
tmp1=pd.merge(finalData,dataStations,left_on='start_station', right_on='id')
datos_con_coordenadas=pd.merge(tmp1,dataStations,left_on='end_station', right_on='id')
datos_con_coordenadas  
```

```python
datos_con_coordenadas['distance']=datos_con_coordenadas.apply(lambda x: computeDistance(x['lat_x'],x['lng_x'],x['lat_y'],x['lng_y']), axis=1)
```
